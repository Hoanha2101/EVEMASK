```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
import os
import random
from matplotlib.patches import Rectangle

# Thiết lập matplotlib để hiển thị tiếng Việt
plt.rcParams['font.family'] = ['DejaVu Sans']
plt.style.use('dark_background')

class ImageAugmentation:
    def __init__(self, image_path, overlay_folder):
        self.image_path = image_path
        self.overlay_folder = overlay_folder
        self.original_image = cv2.imread(image_path)
        self.original_image = cv2.cvtColor(self.original_image, cv2.COLOR_BGR2RGB)
        
        # Lấy danh sách ảnh phụ
        if os.path.exists(overlay_folder):
            self.overlay_images = [f for f in os.listdir(overlay_folder) 
                                 if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        else:
            print(f"Thư mục {overlay_folder} không tồn tại!")
            self.overlay_images = []
    
    def change_color_random(self, image):
        """Phương pháp 1: Đổi màu ảnh random"""
        results = []
        
        for i in range(3):
            # Random hue shift
            hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
            hue_shift = random.randint(-50, 50)
            hsv[:, :, 0] = (hsv[:, :, 0] + hue_shift) % 180
            
            # Random saturation
            sat_factor = random.uniform(0.7, 1.3)
            hsv[:, :, 1] = np.clip(hsv[:, :, 1] * sat_factor, 0, 255)
            
            # Random brightness
            val_factor = random.uniform(0.8, 1.2)
            hsv[:, :, 2] = np.clip(hsv[:, :, 2] * val_factor, 0, 255)
            
            result = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
            results.append(result)
        
        return results
    
    def rotate_random(self, image):
        """Phương pháp 2: Xoay ảnh random từ (-90,90) với padding"""
        results = []
        h, w = image.shape[:2]
        center = (w // 2, h // 2)
        
        for i in range(3):
            angle = random.uniform(-90, 90)
            
            # Tính toán kích thước canvas mới để chứa toàn bộ ảnh sau khi xoay
            rad = np.radians(abs(angle))
            new_w = int(h * np.sin(rad) + w * np.cos(rad))
            new_h = int(h * np.cos(rad) + w * np.sin(rad))
            
            # Ma trận xoay với tâm mới
            M = cv2.getRotationMatrix2D(center, angle, 1.0)
            
            # Điều chỉnh translation để ảnh nằm giữa canvas mới
            M[0, 2] += (new_w - w) / 2
            M[1, 2] += (new_h - h) / 2
            
            # Tạo màu padding random
            padding_colors = [
                [0, 0, 0],           # Đen
                [255, 255, 255],     # Trắng
                [128, 128, 128],     # Xám
                [64, 64, 64],        # Xám đậm
                [192, 192, 192],     # Xám nhạt
                [255, 240, 245],     # Hồng nhạt
                [240, 248, 255],     # Xanh nhạt
                [255, 248, 220],     # Vàng nhạt
                [248, 248, 255],     # Lavender
                [245, 245, 220],     # Beige
            ]
            
            background_color = random.choice(padding_colors)
            
            # Xoay ảnh với canvas mới
            rotated = cv2.warpAffine(image, M, (new_w, new_h), 
                                   borderMode=cv2.BORDER_CONSTANT,
                                   borderValue=background_color)
            
            print(f"Xoay ảnh {i+1}: góc {angle:.1f}°, kích thước {w}x{h} → {new_w}x{new_h}, màu padding {background_color}")
            
            results.append(rotated)
        
        return results
    
    def add_overlay(self, image):
        """Phương pháp 3: Chèn ảnh phụ PNG (1/6 diện tích)"""
        results = []
        
        if not self.overlay_images:
            print("Không có ảnh phụ để chèn!")
            return [image] * 3
        
        print(f"Tìm thấy {len(self.overlay_images)} ảnh trong folder cau_thu: {self.overlay_images}")
        
        h, w = image.shape[:2]
        image_area = h * w
        
        # Tính kích thước ảnh phụ để có diện tích = 1/6 diện tích ảnh gốc
        overlay_area = image_area / 6
        overlay_size = int(np.sqrt(overlay_area))  # Giả sử ảnh phụ vuông
        
        print(f"Kích thước ảnh gốc: {w}x{h}")
        print(f"Diện tích ảnh gốc: {image_area:,} pixels")
        print(f"Diện tích ảnh phụ (1/6): {overlay_area:,} pixels")
        print(f"Kích thước ảnh phụ: {overlay_size}x{overlay_size}")
        
        for i in range(3):
            result = image.copy()
            
            # Chọn ảnh phụ random từ 4 ảnh
            overlay_name = random.choice(self.overlay_images)
            overlay_path = os.path.join(self.overlay_folder, overlay_name)
            print(f"Ảnh {i+1}: Chọn {overlay_name}")
            
            try:
                # Đọc ảnh PNG với alpha channel
                overlay = cv2.imread(overlay_path, cv2.IMREAD_UNCHANGED)
                if overlay is None:
                    print(f"Không thể đọc ảnh {overlay_name}")
                    continue
                
                # Kiểm tra xem ảnh có alpha channel không
                if overlay.shape[2] == 4:  # RGBA
                    # Tách alpha channel
                    overlay_bgr = overlay[:, :, :3]
                    alpha_channel = overlay[:, :, 3]
                    
                    # Chuyển sang RGB
                    overlay_rgb = cv2.cvtColor(overlay_bgr, cv2.COLOR_BGR2RGB)
                    
                    # Resize ảnh phụ và alpha channel
                    overlay_rgb = cv2.resize(overlay_rgb, (overlay_size, overlay_size))
                    alpha_channel = cv2.resize(alpha_channel, (overlay_size, overlay_size))
                    
                    # Chọn vị trí random để chèn ảnh phụ
                    max_x = max(0, w - overlay_size)
                    max_y = max(0, h - overlay_size)
                    
                    if max_x > 0 and max_y > 0:
                        x = random.randint(0, max_x)
                        y = random.randint(0, max_y)
                        
                        # Lấy vùng ROI từ ảnh gốc
                        roi = result[y:y+overlay_size, x:x+overlay_size]
                        
                        # Chuyển alpha channel thành mask (0-1)
                        alpha_mask = alpha_channel.astype(float) / 255.0
                        
                        # Áp dụng alpha blending cho từng channel
                        for c in range(3):
                            roi[:, :, c] = (alpha_mask * overlay_rgb[:, :, c] + 
                                          (1 - alpha_mask) * roi[:, :, c])
                        
                        # Gán kết quả về ảnh
                        result[y:y+overlay_size, x:x+overlay_size] = roi.astype(np.uint8)
                        
                        print(f"  → Chèn PNG với alpha tại vị trí ({x}, {y})")
                    else:
                        print(f"  → Ảnh phụ quá lớn so với ảnh gốc!")
                
                else:  # RGB thông thường
                    print(f"  → Ảnh {overlay_name} không có alpha channel, xử lý như ảnh thường")
                    overlay = cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)
                    overlay = cv2.resize(overlay, (overlay_size, overlay_size))
                    
                    max_x = max(0, w - overlay_size)
                    max_y = max(0, h - overlay_size)
                    
                    if max_x > 0 and max_y > 0:
                        x = random.randint(0, max_x)
                        y = random.randint(0, max_y)
                        
                        # Alpha blending thông thường
                        alpha = 0.8
                        roi = result[y:y+overlay_size, x:x+overlay_size]
                        blended = cv2.addWeighted(overlay, alpha, roi, 1-alpha, 0)
                        result[y:y+overlay_size, x:x+overlay_size] = blended
                        
                        print(f"  → Chèn ảnh thường tại vị trí ({x}, {y})")
                
            except Exception as e:
                print(f"Lỗi khi chèn ảnh {overlay_name}: {e}")
            
            results.append(result)
        
        return results
    
    def zoom_random(self, image):
        """Phương pháp 4: Zoom-in, zoom-out random"""
        results = []
        h, w = image.shape[:2]
        
        for i in range(3):
            # Random zoom factor (0.7-1.5)
            zoom_factor = random.uniform(0.7, 1.5)
            
            if zoom_factor < 1:  # Zoom out
                new_h, new_w = int(h * zoom_factor), int(w * zoom_factor)
                resized = cv2.resize(image, (new_w, new_h))
                
                # Tạo canvas mới và đặt ảnh vào giữa
                result = np.zeros_like(image)
                start_y = (h - new_h) // 2
                start_x = (w - new_w) // 2
                result[start_y:start_y+new_h, start_x:start_x+new_w] = resized
                
            else:  # Zoom in
                new_h, new_w = int(h * zoom_factor), int(w * zoom_factor)
                resized = cv2.resize(image, (new_w, new_h))
                
                # Crop từ giữa
                start_y = (new_h - h) // 2
                start_x = (new_w - w) // 2
                result = resized[start_y:start_y+h, start_x:start_x+w]
            
            results.append(result)
        
        return results
    
    def skew_random(self, image):
        """Phương pháp 5: Nghiêng ảnh random"""
        results = []
        h, w = image.shape[:2]
        
        for i in range(3):
            # Random skew parameters (giảm độ nghiêng để tránh biến dạng quá mức)
            skew_factor = random.uniform(0.1, 0.3)
            
            # Các điểm góc gốc
            src_points = np.float32([[0, 0], [w-1, 0], [w-1, h-1], [0, h-1]])
            
            # Tạo các điểm đích với độ nghiêng ngẫu nhiên
            skew_type = random.choice(['left', 'right', 'top', 'bottom'])
            
            if skew_type == 'left':  # Nghiêng trái
                dst_points = np.float32([
                    [w*skew_factor, 0], 
                    [w-1, 0], 
                    [w-1, h-1], 
                    [0, h-1]
                ])
            elif skew_type == 'right':  # Nghiêng phải
                dst_points = np.float32([
                    [0, 0], 
                    [w-1-w*skew_factor, 0], 
                    [w-1, h-1], 
                    [w*skew_factor, h-1]
                ])
            elif skew_type == 'top':  # Nghiêng lên
                dst_points = np.float32([
                    [0, h*skew_factor], 
                    [w-1, 0], 
                    [w-1, h-1-h*skew_factor], 
                    [0, h-1]
                ])
            else:  # Nghiêng xuống
                dst_points = np.float32([
                    [0, 0], 
                    [w-1, h*skew_factor], 
                    [w-1, h-1], 
                    [0, h-1-h*skew_factor]
                ])
            
            # Áp dụng biến đổi perspective
            M = cv2.getPerspectiveTransform(src_points, dst_points)
            background_color = [0, 0, 0] if random.random() > 0.5 else [255, 255, 255]
            skewed = cv2.warpPerspective(image, M, (w, h), 
                                       borderMode=cv2.BORDER_CONSTANT,
                                       borderValue=background_color)
            results.append(skewed)
        
        return results
    
    def visualize_all(self):
        """Trực quan hóa tất cả các phương pháp"""
        # Áp dụng các phương pháp
        color_results = self.change_color_random(self.original_image)
        rotate_results = self.rotate_random(self.original_image)
        overlay_results = self.add_overlay(self.original_image)
        zoom_results = self.zoom_random(self.original_image)
        skew_results = self.skew_random(self.original_image)
        
        # Tạo figure lớn
        fig, axes = plt.subplots(6, 4, figsize=(16, 20))
        fig.suptitle('Kết quả xử lý ảnh - 5 phương pháp', fontsize=20, fontweight='bold')
        
        # Hiển thị ảnh gốc
        axes[0, 0].imshow(self.original_image)
        axes[0, 0].set_title('Ảnh gốc', fontsize=14, fontweight='bold')
        axes[0, 0].axis('off')
        
        # Ẩn các ô trống ở hàng đầu
        for j in range(1, 4):
            axes[0, j].axis('off')
        
        # Hiển thị kết quả từng phương pháp
        methods = [
            ('Đổi màu random', color_results),
            ('Xoay ảnh random', rotate_results), 
            ('Chèn ảnh phụ', overlay_results),
            ('Zoom random', zoom_results),
            ('Nghiêng ảnh random', skew_results)
        ]
        
        for i, (method_name, results) in enumerate(methods):
            row = i + 1
            axes[row, 0].text(0.5, 0.5, method_name, 
                            ha='center', va='center', fontsize=12, 
                            fontweight='bold', transform=axes[row, 0].transAxes)
            axes[row, 0].axis('off')
            
            for j, result in enumerate(results):
                if j < 3:  # Chỉ hiển thị 3 ảnh
                    axes[row, j+1].imshow(result)
                    axes[row, j+1].set_title(f'Ảnh {j+1}', fontsize=10)
                    axes[row, j+1].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        return {
            'color_change': color_results,
            'rotation': rotate_results,
            'overlay': overlay_results,
            'zoom': zoom_results,
            'skew': skew_results
        }

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
import os
import random
from matplotlib.patches import Rectangle

# Thiết lập matplotlib để hiển thị tiếng Việt
plt.rcParams['font.family'] = ['DejaVu Sans']
plt.style.use('dark_background')

class ImageAugmentation:
    def __init__(self, image_path, overlay_folder):
        self.image_path = image_path
        self.overlay_folder = overlay_folder
        self.original_image = cv2.imread(image_path)
        self.original_image = cv2.cvtColor(self.original_image, cv2.COLOR_BGR2RGB)
        
        # Lấy danh sách ảnh phụ
        if os.path.exists(overlay_folder):
            self.overlay_images = [f for f in os.listdir(overlay_folder) 
                                 if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        else:
            print(f"Thư mục {overlay_folder} không tồn tại!")
            self.overlay_images = []
    
    def change_color_random(self, image):
        """Phương pháp 1: Đổi màu ảnh random"""
        results = []
        
        for i in range(3):
            # Random hue shift
            hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
            hue_shift = random.randint(-50, 50)
            hsv[:, :, 0] = (hsv[:, :, 0] + hue_shift) % 180
            
            # Random saturation
            sat_factor = random.uniform(0.7, 1.3)
            hsv[:, :, 1] = np.clip(hsv[:, :, 1] * sat_factor, 0, 255)
            
            # Random brightness
            val_factor = random.uniform(0.8, 1.2)
            hsv[:, :, 2] = np.clip(hsv[:, :, 2] * val_factor, 0, 255)
            
            result = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
            results.append(result)
        
        return results
    
    def rotate_random(self, image):
        """Phương pháp 2: Xoay ảnh random từ (-90,90) với padding"""
        results = []
        h, w = image.shape[:2]
        center = (w // 2, h // 2)
        
        for i in range(3):
            angle = random.uniform(-90, 90)
            
            # Tính toán kích thước canvas mới để chứa toàn bộ ảnh sau khi xoay
            rad = np.radians(abs(angle))
            new_w = int(h * np.sin(rad) + w * np.cos(rad))
            new_h = int(h * np.cos(rad) + w * np.sin(rad))
            
            # Ma trận xoay với tâm mới
            M = cv2.getRotationMatrix2D(center, angle, 1.0)
            
            # Điều chỉnh translation để ảnh nằm giữa canvas mới
            M[0, 2] += (new_w - w) / 2
            M[1, 2] += (new_h - h) / 2
            
            # Tạo màu padding random
            padding_colors = [
                [0, 0, 0],           # Đen
                [255, 255, 255],     # Trắng
                [128, 128, 128],     # Xám
                [64, 64, 64],        # Xám đậm
                [192, 192, 192],     # Xám nhạt
                [255, 240, 245],     # Hồng nhạt
                [240, 248, 255],     # Xanh nhạt
                [255, 248, 220],     # Vàng nhạt
                [248, 248, 255],     # Lavender
                [245, 245, 220],     # Beige
            ]
            
            background_color = random.choice(padding_colors)
            
            # Xoay ảnh với canvas mới
            rotated = cv2.warpAffine(image, M, (new_w, new_h), 
                                   borderMode=cv2.BORDER_CONSTANT,
                                   borderValue=background_color)
            
            print(f"Xoay ảnh {i+1}: góc {angle:.1f}°, kích thước {w}x{h} → {new_w}x{new_h}, màu padding {background_color}")
            
            results.append(rotated)
        
        return results
    
    def add_overlay(self, image):
        """Phương pháp 3: Chèn ảnh phụ PNG (1/6 diện tích)"""
        results = []
        
        if not self.overlay_images:
            print("Không có ảnh phụ để chèn!")
            return [image] * 3
        
        print(f"Tìm thấy {len(self.overlay_images)} ảnh trong folder cau_thu: {self.overlay_images}")
        
        h, w = image.shape[:2]
        image_area = h * w
        
        # Tính kích thước ảnh phụ để có diện tích = 1/6 diện tích ảnh gốc
        overlay_area = image_area / 6
        overlay_size = int(np.sqrt(overlay_area))  # Giả sử ảnh phụ vuông
        
        print(f"Kích thước ảnh gốc: {w}x{h}")
        print(f"Diện tích ảnh gốc: {image_area:,} pixels")
        print(f"Diện tích ảnh phụ (1/6): {overlay_area:,} pixels")
        print(f"Kích thước ảnh phụ: {overlay_size}x{overlay_size}")
        
        for i in range(3):
            result = image.copy()
            
            # Chọn ảnh phụ random từ 4 ảnh
            overlay_name = random.choice(self.overlay_images)
            overlay_path = os.path.join(self.overlay_folder, overlay_name)
            print(f"Ảnh {i+1}: Chọn {overlay_name}")
            
            try:
                # Đọc ảnh PNG với alpha channel
                overlay = cv2.imread(overlay_path, cv2.IMREAD_UNCHANGED)
                if overlay is None:
                    print(f"Không thể đọc ảnh {overlay_name}")
                    continue
                
                # Kiểm tra xem ảnh có alpha channel không
                if overlay.shape[2] == 4:  # RGBA
                    # Tách alpha channel
                    overlay_bgr = overlay[:, :, :3]
                    alpha_channel = overlay[:, :, 3]
                    
                    # Chuyển sang RGB
                    overlay_rgb = cv2.cvtColor(overlay_bgr, cv2.COLOR_BGR2RGB)
                    
                    # Resize ảnh phụ và alpha channel
                    overlay_rgb = cv2.resize(overlay_rgb, (overlay_size, overlay_size))
                    alpha_channel = cv2.resize(alpha_channel, (overlay_size, overlay_size))
                    
                    # Chọn vị trí random để chèn ảnh phụ
                    max_x = max(0, w - overlay_size)
                    max_y = max(0, h - overlay_size)
                    
                    if max_x > 0 and max_y > 0:
                        x = random.randint(0, max_x)
                        y = random.randint(0, max_y)
                        
                        # Lấy vùng ROI từ ảnh gốc
                        roi = result[y:y+overlay_size, x:x+overlay_size]
                        
                        # Chuyển alpha channel thành mask (0-1)
                        alpha_mask = alpha_channel.astype(float) / 255.0
                        
                        # Áp dụng alpha blending cho từng channel
                        for c in range(3):
                            roi[:, :, c] = (alpha_mask * overlay_rgb[:, :, c] + 
                                          (1 - alpha_mask) * roi[:, :, c])
                        
                        # Gán kết quả về ảnh
                        result[y:y+overlay_size, x:x+overlay_size] = roi.astype(np.uint8)
                        
                        print(f"  → Chèn PNG với alpha tại vị trí ({x}, {y})")
                    else:
                        print(f"  → Ảnh phụ quá lớn so với ảnh gốc!")
                
                else:  # RGB thông thường
                    print(f"  → Ảnh {overlay_name} không có alpha channel, xử lý như ảnh thường")
                    overlay = cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)
                    overlay = cv2.resize(overlay, (overlay_size, overlay_size))
                    
                    max_x = max(0, w - overlay_size)
                    max_y = max(0, h - overlay_size)
                    
                    if max_x > 0 and max_y > 0:
                        x = random.randint(0, max_x)
                        y = random.randint(0, max_y)
                        
                        # Alpha blending thông thường
                        alpha = 0.8
                        roi = result[y:y+overlay_size, x:x+overlay_size]
                        blended = cv2.addWeighted(overlay, alpha, roi, 1-alpha, 0)
                        result[y:y+overlay_size, x:x+overlay_size] = blended
                        
                        print(f"  → Chèn ảnh thường tại vị trí ({x}, {y})")
                
            except Exception as e:
                print(f"Lỗi khi chèn ảnh {overlay_name}: {e}")
            
            results.append(result)
        
        return results
    
    def zoom_random(self, image):
        """Phương pháp 4: Zoom-in, zoom-out random"""
        results = []
        h, w = image.shape[:2]
        
        for i in range(3):
            # Random zoom factor (0.7-1.5)
            zoom_factor = random.uniform(0.7, 1.5)
            
            if zoom_factor < 1:  # Zoom out
                new_h, new_w = int(h * zoom_factor), int(w * zoom_factor)
                resized = cv2.resize(image, (new_w, new_h))
                
                # Tạo canvas mới và đặt ảnh vào giữa
                result = np.zeros_like(image)
                start_y = (h - new_h) // 2
                start_x = (w - new_w) // 2
                result[start_y:start_y+new_h, start_x:start_x+new_w] = resized
                
            else:  # Zoom in
                new_h, new_w = int(h * zoom_factor), int(w * zoom_factor)
                resized = cv2.resize(image, (new_w, new_h))
                
                # Crop từ giữa
                start_y = (new_h - h) // 2
                start_x = (new_w - w) // 2
                result = resized[start_y:start_y+h, start_x:start_x+w]
            
            results.append(result)
        
        return results
    
    def skew_random(self, image):
        """Phương pháp 5: Nghiêng ảnh random"""
        results = []
        h, w = image.shape[:2]
        
        for i in range(3):
            # Random skew parameters (giảm độ nghiêng để tránh biến dạng quá mức)
            skew_factor = random.uniform(0.1, 0.3)
            
            # Các điểm góc gốc
            src_points = np.float32([[0, 0], [w-1, 0], [w-1, h-1], [0, h-1]])
            
            # Tạo các điểm đích với độ nghiêng ngẫu nhiên
            skew_type = random.choice(['left', 'right', 'top', 'bottom'])
            
            if skew_type == 'left':  # Nghiêng trái
                dst_points = np.float32([
                    [w*skew_factor, 0], 
                    [w-1, 0], 
                    [w-1, h-1], 
                    [0, h-1]
                ])
            elif skew_type == 'right':  # Nghiêng phải
                dst_points = np.float32([
                    [0, 0], 
                    [w-1-w*skew_factor, 0], 
                    [w-1, h-1], 
                    [w*skew_factor, h-1]
                ])
            elif skew_type == 'top':  # Nghiêng lên
                dst_points = np.float32([
                    [0, h*skew_factor], 
                    [w-1, 0], 
                    [w-1, h-1-h*skew_factor], 
                    [0, h-1]
                ])
            else:  # Nghiêng xuống
                dst_points = np.float32([
                    [0, 0], 
                    [w-1, h*skew_factor], 
                    [w-1, h-1], 
                    [0, h-1-h*skew_factor]
                ])
            
            # Áp dụng biến đổi perspective
            M = cv2.getPerspectiveTransform(src_points, dst_points)
            background_color = [0, 0, 0] if random.random() > 0.5 else [255, 255, 255]
            skewed = cv2.warpPerspective(image, M, (w, h), 
                                       borderMode=cv2.BORDER_CONSTANT,
                                       borderValue=background_color)
            results.append(skewed)
        
        return results
    
    def visualize_all(self):
        """Trực quan hóa tất cả các phương pháp"""
        # Áp dụng các phương pháp
        color_results = self.change_color_random(self.original_image)
        rotate_results = self.rotate_random(self.original_image)
        overlay_results = self.add_overlay(self.original_image)
        zoom_results = self.zoom_random(self.original_image)
        skew_results = self.skew_random(self.original_image)
        
        # Tạo figure lớn
        fig, axes = plt.subplots(6, 4, figsize=(16, 20))
        fig.suptitle('Kết quả xử lý ảnh - 5 phương pháp', fontsize=20, fontweight='bold')
        
        # Hiển thị ảnh gốc
        axes[0, 0].imshow(self.original_image)
        axes[0, 0].set_title('Ảnh gốc', fontsize=14, fontweight='bold')
        axes[0, 0].axis('off')
        
        # Ẩn các ô trống ở hàng đầu
        for j in range(1, 4):
            axes[0, j].axis('off')
        
        # Hiển thị kết quả từng phương pháp
        methods = [
            ('Đổi màu random', color_results),
            ('Xoay ảnh random', rotate_results), 
            ('Chèn ảnh phụ', overlay_results),
            ('Zoom random', zoom_results),
            ('Nghiêng ảnh random', skew_results)
        ]
        
        for i, (method_name, results) in enumerate(methods):
            row = i + 1
            axes[row, 0].text(0.5, 0.5, method_name, 
                            ha='center', va='center', fontsize=12, 
                            fontweight='bold', transform=axes[row, 0].transAxes)
            axes[row, 0].axis('off')
            
            for j, result in enumerate(results):
                if j < 3:  # Chỉ hiển thị 3 ảnh
                    axes[row, j+1].imshow(result)
                    axes[row, j+1].set_title(f'Ảnh {j+1}', fontsize=10)
                    axes[row, j+1].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        return {
            'color_change': color_results,
            'rotation': rotate_results,
            'overlay': overlay_results,
            'zoom': zoom_results,
            'skew': skew_results
        }

# Sử dụng
if __name__ == "__main__":
    # Khởi tạo
    image_path = "3M.jpg"  # Đường dẫn đến ảnh gốc
    overlay_folder = "cau_thu"  # Thư mục chứa ảnh phụ
    
    # Kiểm tra file tồn tại
    if not os.path.exists(image_path):
        print(f"Không tìm thấy file ảnh: {image_path}")
        print("Vui lòng đảm bảo file ảnh tồn tại!")
    else:
        # Tạo đối tượng xử lý ảnh
        aug = ImageAugmentation(image_path, overlay_folder)
        
        # Thực hiện và hiển thị kết quả
        results = aug.visualize_all()
        
        # In thông tin
        print("\n" + "="*50)
        print("HOÀN THÀNH XỬ LÝ ảnh!")
        print("="*50)
        print(f"✓ Đã tạo 3 ảnh đổi màu random")
        print(f"✓ Đã tạo 3 ảnh xoay random (-90° đến 90°)")
        print(f"✓ Đã tạo 3 ảnh chèn ảnh phụ (1/5 diện tích)")
        print(f"✓ Đã tạo 3 ảnh zoom random")
        print(f"✓ Đã tạo 3 ảnh nghiêng random")
        print(f"✓ Tổng cộng: 15 ảnh đã được tạo")
```
