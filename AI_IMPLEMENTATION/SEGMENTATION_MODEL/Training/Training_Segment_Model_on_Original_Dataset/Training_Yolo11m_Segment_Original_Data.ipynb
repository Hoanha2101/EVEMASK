```python
from ultralytics import YOLO
def main():
    model = YOLO('/kaggle/input/load-model/yolo11m-seg.pt')  
    results = model.train(data='/kaggle/input/file-yyaml-new/dataset.yaml',imgsz=640,epochs=100,
                          optimizer="AdamW",  batch=16, device=0, patience=1000,
                          lr0 = 0.01,
                          fliplr = 0, crop_fraction = 0 ,mixup =0, copy_paste=1,perspective=0,
                          mosaic=0,shear=0,degrees=0,
                          hsv_h=0.15, hsv_s=0.7, hsv_v = 0.5, translate = 0)
                         
if __name__ == '__main__':
    main()
```
